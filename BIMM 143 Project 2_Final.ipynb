### Scientific Question: Are Per1 and Per2Â proteins in Arvicanthis niloticus extremely different from Mus musculus given that they have different circadian rhythms?

Both Per1 and Per2 proteins play an important role in the circadian rhythm [1]. Circadian rhythm is an internal biological process that helps to regulate the sleep-wake cycle, which repeats roughly every 24 hours [2]. If Per1 and Per2 were to be deleted from an animal, then the function of its internal clock with be destroyed, causing an abnormal circadian pattern [3].


Arvicanthis niloticus common name is African grass rat [4]. These mice are diurnal, which means they are active during the daytime. African grass rats generally live in various habitats such as desert, grassland, and scrub forest. Furthermore, these African grass rats are primarily herbivores that eat leaves, wood, seeds, and nuts. Mus musculus common name is house mice are usually nocturnal, which means they are active during the night [5]. House mice, like its name, generally live in places that are closed to humans, such as houses and barns. They tend to stay away from urban locations. House mice are omnivores. Thus, they have a diet of eating things from carrion and insects to seeds and nuts. 


UniProt stands for Universal Protein Resource [6]. It is a publicly accessible database that contains protein sequences and annotation data. There are three databases in UniProt, which are UniProt Knowledgebase (UniProtKB), the UniProt Reference Clusters (UniRef), and the UniProt Archive (UniParc). Furthermore, UniProt is a collaboration between three institutions that are committed to the preservation of the UniProt databases. The three institutions are European Bioinformatics Institute (EMBL-EBI), the SIB Swiss Institute of Bioinformatics, and the Protein Information Resource (PIR).

```python
#Created a function for pairwise sequences
def pairwise(seq1, seq2):
    #Define a local variable alignment
    alignments = pairwise2.align.globalxx(seq1, seq2)
    for match in alignments:
        #Define a global variable my_pairwise
        my_pairwise = print(match)
    return my_pairwise
```

```python
#pairwise sequences between Per1 of African grass rat and Per 1 Mus musculus
pairwise(Per1_grass, Per1_mice)
```

```python
#pairwise sequences between Per2 of African grass rat and Per 2 Mus musculus
pairwise(Per2_grass, Per2_mice)
```

```python
#Creating a function for multiple sequence alignment
def multiple_align(FASTA_files):
    #Define a local variable input_file
    input_file = FASTA_files
    #Define a local variable records
    records = SeqIO.parse(input_file, 'fasta')
    records = list(records) 
    #Define a local variable maxlen
    maxlen = max(len(record.seq) for record in records)

    #This part of the code allows all the length of all the protein sequences in the FASTA file to have the same length
    for record in records:
        if len(record.seq) != maxlen:
            #Define a local variable sequence
            sequence = str(record.seq).ljust(maxlen, '.')
            #Define a local variable record.seq
            record.seq = Seq.Seq(sequence)
    assert all(len(record.seq) == maxlen for record in records)
    
    #Creating an file to check everything is working correctly
    #Define a local variable output_file
    output_file = '{}_padded.fasta'.format(os.path.splitext(input_file)[0])
    with open(output_file, 'w') as f:
        SeqIO.write(records, f, 'fasta')
        
    #Now that the preparation is set up, multiple sequence alignment code is used
    #Define a local variable alignment
    alignment = AlignIO.read(output_file, "fasta")
    return print(alignment)
```

### Part 3: Create a heatmap showing the pairwise sequences for Per1 and Per2 of mice

Here I used SeqIO, numpy, and matplotlib to create a heatmap. I used SeqIO to make my code to read my FASTA file called, "experiment_per1.fasta.txt". I used numpy to create a pairwise_seq_heatmap function that would give me a 6X6 numpy array of how similar my FASTA files are with one another. Next, I used matplotlib to create a plot, which the x and y axes are label by the type of animal. The range of the heatmap is from 1.0 to 0.0. Heatmaps are a good visualization for see similarity between data using a gradient color, which is shown below. 

```python
#Making Per1 FASTA file into a list
#Define a global variable Fasta
Fasta = list(SeqIO.parse("experiment_per1.fasta.txt", "fasta"))

#Making a code for the length of the Per1 FASTA file
#Define a global variable size
size = len(Fasta)
#print(size) --> It should be 6 when you print it.

#Initiate variable i and j
#i = 0
#j = 0

#Making an emtpy numpy matrix that is the length of the fasta file
#Define a global variable alignments_score
alignments_score = np.empty(shape=(size,size))
```

```python
#Creating a function that 
def pairwise_seq_heatmap(my_numpy_array, size):
    for i in range(size):
        for j in range(size):
        #Define two sequences to be aligned
        #Define a local variable X
        #Define a local variable Y
            X = Fasta[i].seq
            Y = Fasta[j].seq

        # Get a list of the global alignments between the two sequences ACGGGT and ACG
        # No parameters. Identical characters have score of 1, else 0.
        # No gap penalties.
        #Define a local variable my_numpy_array[i,j]
            my_numpy_array[i,j] = int(pairwise2.align.globalxx(X, Y, score_only=True))
    
    #Define a local variable empty array
    empty_array =[[]]
    #Define a local variable c
    c=0
    for i in my_numpy_array: 
        #Define a local variable a
        a= np.divide(i,i[c])
        empty_array= np.append(empty_array,a)
        c+=1  
    #print(empty_array.shape) -> It should come out to be (36,)
    
    #Reshaping the empty array to make it (6,6)
    #Define a local variable empty_array 2
    empty_array_2 = np.reshape(empty_array, (-1, 6))
    
    #Transpose the empty_array_2 back since the my_numpy_array got transpose when dividing and appending to empty_array
    empty_array_2= (np.transpose(empty_array_2))
    
    #Setting the numpy array to 3 decimal places
    np.set_printoptions(precision=3)
    return empty_array_2
    #print(empty_array_2.shape) -> Making sure it is (6,6) and not (36,)
    #print(empty_array_2) -> The numpy array should be between 1.0 to 0.0. It the numbers are greater than 1 or less than 0, then there is an error. 
    
#Define a local variable My_alignment_Score
My_alignment_Score = pairwise_seq_heatmap(alignments_score, size)
#print(My_alignment_Score)
```

```python
#Define a global variable fig, ax
fig, ax = plt.subplots()
#Define a global variable seqs
seqs = ['grass mice','human','dog','mice','bat','rat']
plt.imshow(My_alignment_Score)
#Define a global variable im
im = ax.imshow(My_alignment_Score)


#Labeling the x and y axis
ax.set_xticks(np.arange(len(seqs)))
ax.set_yticks(np.arange(len(seqs)))
ax.set_xticklabels(seqs)
ax.set_yticklabels(seqs)

#Label the title
plt.title('PER1 Heatmap')

#Define a global variable cbar
cbar = ax.figure.colorbar(im)
plt.show()
```

```python
#Define a global variable Fasta
Fasta = list(SeqIO.parse("experiment_per2.fasta.txt", "fasta"))
#Define a global variable size
size = len(Fasta)
#Define a global variable alignments_score
alignments_score = np.empty(shape=(size,size))
```

```python
#Define a global variable My_alignment_Scores
My_alignment_Scores = pairwise_seq_heatmap(alignments_score, size)
```

```python
#Create subplot
#Define a global variable fig, ax
fig, ax = plt.subplots()

# Actually show the heatplot
plt.imshow(My_alignment_Scores)

# Construct the heat map to show the axes
#Define a global variable im
im = ax.imshow(My_alignment_Scores)

#Labeling the x and y axis
ax.set_xticks(np.arange(len(seqs)))
ax.set_yticks(np.arange(len(seqs)))
ax.set_xticklabels(seqs)
ax.set_yticklabels(seqs)

#Label the title
plt.title('PER2 Heatmap')

# Add a colorbar to the right of the heatmap
#Define a global variable cbar
cbar = ax.figure.colorbar(im)

# Show the plot
plt.show()
```
